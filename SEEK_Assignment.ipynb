```python
def special_sum(lst):
    output_list = []
    for i, position in enumerate(lst):
        total = 0
        for j, number in enumerate(lst):
            if j == i:
                continue
            else:
                total = total + number
        output_list.append(total)
    return output_list
```

## Identify outliers from the data and output the corresponding week numbers

```python
# Outliers
# you can change quantile values to get the result more precise, generally 75% & 25% is used,
# I am using 90% & 10 % quantile
def outliers(x):
   return np.abs(x- x.median()) > 1.5*(x.quantile(.75)-x.quantile(0.25))

def get_outliers_per_week(req_df):
    column_names = req_df.columns.tolist()
    for week in column_names:
        print(f"\nOutliers for {week} : ", req_df[week][outliers(req_df[week])].to_dict())
```

```python
req_df = df_Product_Code_index_T[:52]
get_outliers_per_week(req_df)
```

# Q4.String similarity (code in python):


```python
from difflib import SequenceMatcher


import pandas as pd 
from IPython.display import display, HTML
import re


df_train = pd.read_csv('train.csv', encoding="ISO-8859-1")
df_test = pd.read_csv('test.csv', encoding="ISO-8859-1")


def seq_match(test1, test2):
    return SequenceMatcher(None, test1, test2).ratio()
    
def seq_match_df(df_data):
    col1 = 'description_x'
    col2 = 'description_y'
    df_data[col1] = df_data[col1].str.replace(r'\d', '')
    df_data[col2] = df_data[col2].str.replace(r'\d', '')
    
    df_data['seq_match'] = 0
    df_data['seq_match'] = df_data.apply(
        lambda x: seq_match(x[col1], x[col2]), axis=1)
    
    return df_data

df_result = seq_match_df(df_test)
df_result['same_security'] = df_result['seq_match'] > 0.5    
df_group = df_result.groupby('same_security').size().reset_index()
display(df_result)
df_result.to_csv("String_Similarity.csv",index=False)
```
